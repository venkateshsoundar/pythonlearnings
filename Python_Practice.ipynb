```python
def first_non_repeating(s: str) -> str:
    freq = [0] * 26  # O(1) space

    # Step 1: Count frequency
    for ch in s:
        freq[ord(ch) - ord('a')] += 1

    # Step 2: Find first non-repeating character
    for ch in s:
        if freq[ord(ch) - ord('a')] == 1:
            return ch

    return '$'
```

```python
def find_equilibrium_index(arr: list[int]) -> int:
    # Write your code here
    for i in range(1, len(arr)):
        right_sum = 0
        left_sum = 0
        for j in range(i + 1, len(arr)):
            right_sum = right_sum + arr[j]        

        for k in range(0, i):
            left_sum = left_sum + arr[k]        

        if right_sum == left_sum:
            return i
        
        
    return -1


def find_equilibrium_index(arr: list[int]) -> int:
    # Step 1: Compute total sum of the array
    total_sum = sum(arr)
    left_sum = 0

    # Step 2: Traverse through the array
    for i, num in enumerate(arr):
        # Step 3: Check equilibrium condition
        if left_sum == total_sum - left_sum - num:
            return i  # Found equilibrium index
        # Step 4: Update left_sum
        left_sum += num

    # Step 5: No equilibrium index found
    return -1



arr = [1, 2, 0, 3]
print(find_equilibrium_index(arr))
arr=[-7, 1, 5, 2, -4, 3, 0]
print(find_equilibrium_index(arr))
arr=[0, 0, 0, 0]
print(find_equilibrium_index(arr))
```

```python
import pandas as pd

orders_url = "https://raw.githubusercontent.com/ankitbansal6/namastesql_etl_datasets/refs/heads/main/ta/orders.csv"
products_url = "https://raw.githubusercontent.com/ankitbansal6/namastesql_etl_datasets/refs/heads/main/ta/products.csv"

# use pd.read_csv(url) to read files
df1= pd.read_csv(orders_url)
df1.head(5)
df1['quantity']=df1['quantity'].fillna(0)
```

```python
df1.isnull().sum()
```

```python
df2=pd.read_csv(products_url)
df2.head(5)
```

```python
df2.isnull().sum()
```

```python
df2['category']=df2['category'].fillna('Unknown')
```

```python
df3=pd.merge(df1, df2, how='inner', left_on='product_id', right_on='product_id')
```

```python
df3['revenue']= df3['quantity'] * df3['price']
df3.head()  
```

```python
agg_df = (
    df3.groupby("category", as_index=False)["revenue"]
    .sum()
    .rename(columns={"revenue": "total_revenue"})
)
```

```python
df4 = (
    df3.groupby('category', as_index=False)['revenue']
    .sum()
    .sort_values('revenue', ascending=False)
    .rename(columns={'revenue': 'total_revenue'})
)
display(df4)
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys



#
# Complete the 'transformSentence' function below.
#
# The function is expected to return a STRING.
# The function accepts STRING sentence as parameter.
#

def transformSentence(sentence):
    words = sentence.split()
    result_words = []

    for word in words:
        chars = list(word)   # convert string → list
        
        for i in range(1, len(chars)):
            prev=chars[i-1].lower()
            curr=chars[i].lower()
            if prev < curr:
                chars[i] = chars[i].upper()
            else:
                chars[i] = chars[i].lower()
        
        result_words.append("".join(chars)) 

    return " ".join(result_words)


if __name__ == '__main__':
    

    sentence = input()

    result = transformSentence(sentence)

    print(result)

```
