````python
# Level 1 (Grandparent)
class BaseTest:
    def __init__(self):
        print("BaseTest init")

    def setup(self):
        print("Base setup")

# Level 2 (Parent)
class BrowserSetup(BaseTest):
    def __init__(self):
        super().__init__()           # calls BaseTest __init__
        print("BrowserSetup init")

    def launch_browser(self):
        print("Browser launched")

# Another parent class (for multiple inheritance)
class Logger:
    def __init__(self):
        print("Logger init")

    def log(self, message):
        print(f"LOG: {message}")

# Level 3 (Child)
class LoginTest(BrowserSetup, Logger):
    def __init__(self):
        super().__init__()           # follows MRO
        print("LoginTest init")

    def test_login(self):
        self.setup()                 # from BaseTest
        self.launch_browser()        # from BrowserSetup
        self.log("Login test started")  # from Logger
        print("Login test executed")

# Create object
test = LoginTest()
test.test_login()
print(LoginTest.mro())
````

````python
from collections.abc import Iterable

class Test:
    def add(self, *args):
        def flatten_and_sum(x):
            if isinstance(x, (int, float)) and not isinstance(x, bool):
                return x

            if isinstance(x, Iterable) and not isinstance(x, (str, bytes, dict)):
                total = 0
                for item in x:
                    total += flatten_and_sum(item)
                return total

            if isinstance(x, dict):
                total = 0
                for v in x.values():
                    total += flatten_and_sum(v)
                return total

            return 0

        return flatten_and_sum(args)
    
t = Test()
a = [1, 2, [3, 4], (5, 6), {7, 8}, {'a': 9, 'b': 10}]
t.add(a)
````
